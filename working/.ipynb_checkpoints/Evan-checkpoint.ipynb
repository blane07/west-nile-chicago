```python
# ! We may want to include codesum in final model
# ! Refer to pdf in 'assets' for codesum information

# Not using codesum for this benchmark
wx = weather.drop('CodeSum', axis=1)
wx = wx.drop(columns=['Water1', 'Depart', 'Depth', 'SnowFall', 'Sunrise', 'Sunset'])

# Split station 1 and 2 and join horizontally
wx1 = wx[wx['Station']==1]
wx2 = wx[wx['Station']==2]
wx1 = wx1.drop('Station', axis=1)
wx2 = wx2.drop('Station', axis=1)
wx = wx1.merge(wx2, on='Date', suffixes=('_1', '_2'))

# replace some missing values and T with nan
wx = wx.replace('M', np.nan)
wx = wx.replace('-', np.nan)
wx = wx.replace('T', np.nan)
wx = wx.replace(' T', np.nan)
wx = wx.replace('  T', np.nan)
# ! We may want to include codesum in final model
# ! Refer to pdf in 'assets' for codesum information

# Not using codesum for this benchmark
# wx = weather.drop('CodeSum', axis=1)
```

```python
# Functions to extract month and day from dataset
# You can also use parse_dates of Pandas.
def create_year(x):
    return(x.split('-')[0])

def create_month(x):
    return(x.split('-')[1])

def create_day(x):
    return(x.split('-')[2])

train['month'] = train.Date.apply(create_month)
train['day'] = train.Date.apply(create_day)
train['year'] = train.Date.apply(create_year)

test['month'] = test.Date.apply(create_month)
test['day'] = test.Date.apply(create_day)
test['year'] = test.Date.apply(create_year)
```

```python
# drop address columns
train = train.drop(['Address', 'AddressNumberAndStreet', 'NumMosquitos'], axis = 1)
test = test.drop(['Id', 'Address', 'AddressNumberAndStreet'], axis = 1)
```

```python
# Pipeline for heat maps
d = train.groupby(by=['Latitude', 'Longitude'])['WnvPresent']
```

```python
out = []
counts = d.count().to_dict()
successes = d.sum().to_dict()
for key in counts.keys():
    if counts[key] != 0:
        out += [list(key) + [successes[key] / counts[key]]]
m = max([x[2] for x in out])
for x in out:
    x[2] /= m
```

```python
lat_long_dict = {'x':[x[0] for x in out], 
                 'y':[x[1] for x in out], 
                 'c':[cm.hot(x[2]) for x in out]}
plt.scatter(**lat_long_dict);
```

```python
def encode_cols(data, columns, encoders, fit=False):
    out = data.drop(columns=columns).values.transpose().tolist()
    for i, column in enumerate(columns):
        if fit:
            encoders[i].fit(data[column].values.reshape(-1, 1))
        out += encoders[i].transform(data[column].values.reshape(-1, 1)).transpose().toarray().tolist()
#     temp_df = data.drop(columns=columns)
    return np.array(out).transpose().astype(np.float)
```

```python
X = train.drop(columns=['WnvPresent'])
y = train.WnvPresent.values
```

```python
cols_to_encode = ['Species', 'Trap']
encoders = [OneHotEncoder(handle_unknown='ignore') for _ in range(len(cols_to_encode))]
X = encode_cols(X, cols_to_encode, encoders, True)
X2 = encode_cols(test, cols_to_encode, encoders, False)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
1 - (train.WnvPresent.sum() / train.WnvPresent.shape[0])
```

```python
ss = StandardScaler()
X_train = ss.fit_transform(X_train)
X_test = ss.transform(X_test)

lr = LogisticRegression(penalty='l2')
lr.fit(X_train, y_train)
lr.score(X_test, y_test)
```

```python
ss = StandardScaler()
Xs = ss.fit_transform(X)

cv = StratifiedKFold(n_splits=5, shuffle=True)
lr = LogisticRegression(penalty='l2')
s = cross_val_score(lr, Xs, y, cv=cv)

print('{} +/- {}'.format(s.mean(), s.std()))
```
